```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
years = [2018, 2019, 2020, 2021, 2022]
tea_production = [100, 200, 300, 400, 500]
```

```python
plt.plot(years, tea_production) # plt.plot x, y
```

```python
plt.plot(years, tea_production)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
tea_production = [100, 200, 300, 400, 500]
tea_plantation_area = [1000, 1100, 1200, 1300, 1400]
```

```python
plt.plot(years, tea_production)
plt.plot(years, tea_plantation_area)
plt.xlabel('Years')
plt.ylabel('Amount');
```

# DAY 2

**Tas 1**

- whitegrid

```python
sns.set_style("whitegrid")
```

```python
plt.plot(years, tea_production, marker='o')
plt.plot(years, tea_plantation_area, marker='x')

plt.xlabel('Year')
plt.ylabel('Amount')

plt.title("Tea Production and Tea Plantation Area per Year")
plt.legend(['Tea Production (tons)', 'Tea Plantation Area (hectares)'])
```

```python
sns.set_style("darkgrid")
```

```python
plt.plot(years, tea_production, marker='o')
plt.plot(years, tea_plantation_area, marker='x')

plt.xlabel('Year')
plt.ylabel('Amount')

plt.title("Tea Production and Tea Plantation Area per Year")
plt.legend(['Tea Production (tons)', 'Tea Plantation Area (hectares)'])
```

```python
import pandas as pd
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
beam = [50, 25, 42, 30, 50]
ojek = [11, 12, 13, 14, 15]
```

```python
plt.bar(days, beam) # plt.bar(x, y)

plt.xlabel('Day')
plt.ylabel('Transportation (per student)')

plt.title("Use of Public Transportation in Banten Jaya University")
```

```python
# Stacked Bar

plt.bar(days, beam)
plt.bar(days, ojek, bottom=beam)
```
