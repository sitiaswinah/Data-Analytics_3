# Line Chart

Bagan garis adalah grafik yang merepresentasikan informasi sebagai rangkaian titik data yang dihubungkan oleh garis lurus. Dalam diagram garis, setiap titik data atau penanda diplot dan dihubungkan dengan garis atau kurva

Mari kita lihat hasil apel (ton per hektar) di Kanto. Mari kita buat grafik garis menggunakan data ini dan lihat bagaimana hasil panen apel berubah seiring waktu. Kami akan membuat plot dengan menggunakan Matplotlib dan Seaborn sebagai berikut

# Using Matplotlib

Kami menggunakan titik data acak untuk mewakili hasil apel

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton perhectare
```

Setelah kita menentukan titik data hasil panen apel, yang akan kita plot adalah Diagram Garis menggunakan Matplotlib

```python
plt.plot(yield_apples)
```

To better understand the graph and its purpose we can add the x-axis values too

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yeald_appless = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

Karena kita memiliki variabel 'tahun' sebagai sumbu x, kita perlu mengetikkannya sebelum variabel hasil_apel di bawah

```python
plt.plot(years, yield_apples) #plt.plot(x,y)
```

Mari tambahkan label pada sumbu sehingga kita dapat memperlihatkan apa yang diwakili oleh setiap sumbu

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

Untuk memplot beberapa kumpulan data pada grafik yang sama, cukup gunakan fungsi plt.plot satu kali untuk setiap kumpulan data. Mari kita gunakan ini untuk membandingkan hasil apel vs. jeruk pada grafik yang sama

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)');
```

Kita dapat menambahkan legenda yang memberi tahu kita arti setiap baris dalam grafik kita. Untuk memahami apa yang sedang kita rencanakan. kita dapat menambahkan judul ke grafik kita

```python
plt.plot(years, apples)
plt.plot(years, tangerines, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yield (tons per hectare)")
plt.legend(['Apples', 'Tangerines'])
```

# Bar Graph

Bila Anda memiliki data kategorikal, Anda dapat merepresentasikannya dengan grafik batang. Grafik batang memplot data dengan bantuan batang. yang mewakili nilai pada sumbu y dan kategori pada sumbu x. Grafik batang menggunakan batang dengan ketinggian yang bervariasi untuk menunjukkan data yang termasuk dalam kategori tertentu

# Using Seaborn

Mari kita gunakan kumpulan data tips di Seaborn selanjutnya. Kumpulan data terdiri dari:

- informasi tentang jenis kelamin (gender)
- Waktu hari ini
- Jumlah tagihan
- Tips yang diberikan oleh pelanggan yang mengunjungi restoran selama seminggu

Kumpulan **Data Tips** ini adalah kumpulan data yang disediakan oleh perpustakaan seaborn. sehingga Anda dapat langsung memuat dataset

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

Kita dapat menggambar diagram batang untuk memvisualisasikan bagaimana jumlah tagihan rata-rata bervariasi pada hari-hari berbeda dalam seminggu. Kita dapat melakukan ini dengan menghitung rata-rata harian dan kemudian menggunakan plt bar. perpustakaan seaborn juga menyediakan fungsi barplot yang dapat menghitung rata-rata secara otomatis

```python
sns.barplot(x='day',
            y='total_bill',
            data=tips_df)
```

**Vertical Pivot**

```python
sns.barplot(x='day', y='total_bill', hue='sex', data = tips_df)
```

**Horizontal Pivot**

```python
sns.barplot(x='total_bill', y='day', hue='sex', data = tips_df)
```
